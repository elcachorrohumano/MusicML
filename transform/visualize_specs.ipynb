```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

# Path to your HDF5 file
h5_file = '/Users/elcachorrohumano/workspace/MusicNN/data/spectrograms.h5'

# Load the song names and convert byte strings to regular strings
with h5py.File(h5_file, 'r') as f:
    spectrograms = np.array(f['spectrograms'])
    labels = np.array(f['labels'])
    song_names = [name.decode('utf-8') for name in f['song_names']]  # Decode each name from bytes

# Function to visualize a spectrogram
def visualize_spectrogram(spectrogram, label, song_name):
    plt.figure(figsize=(10, 4))
    plt.title(f'Spectrogram for {song_name} (Label: {label})')
    plt.imshow(spectrogram, aspect='auto', origin='lower')
    plt.colorbar(format='%+2.0f dB')
    plt.ylabel('Mel bands')
    plt.xlabel('Time frames')
    plt.tight_layout()
    plt.show()
```

### Spectrogram examples

The Place Where He Inserted the Blade:

```python
# Ask the user for a song name
song_name = "The Place Where He Inserted the Blade"
# Check if the song exists in the dataset
if song_name in song_names:
    index = song_names.index(song_name)  # Find the index of the song using the .index() method
    visualize_spectrogram(spectrograms[index], labels[index], song_name)
else:
    print(f"Song '{song_name}' not found in the dataset.")
```
