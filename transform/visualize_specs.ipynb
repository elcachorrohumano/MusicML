```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

# Path to your HDF5 file
h5_file = '/Users/elcachorrohumano/workspace/MusicNN/data/spectrograms.h5'

# Load the spectrograms and labels from the HDF5 file
with h5py.File(h5_file, 'r') as f:
    spectrograms = np.array(f['spectrograms'])
    labels = np.array(f['labels'])

# Function to visualize spectrograms
def visualize_spectrogram(spectrogram, label, index):
    plt.figure(figsize=(10, 4))
    plt.title(f'Spectrogram (Label: {label}, Index: {index})')
    plt.imshow(spectrogram, aspect='auto', origin='lower')
    plt.colorbar(format='%+2.0f dB')
    plt.ylabel('Mel bands')
    plt.xlabel('Time frames')
    plt.tight_layout()
    plt.show()

# Visualize a few spectrograms
num_spectrograms_to_visualize = 5  # Set how many you want to display
for i in range(num_spectrograms_to_visualize):
    visualize_spectrogram(spectrograms[i], labels[i], i)
```
